```python
from transformers import pipeline

cla = pipeline("sentiment-analysis")
cla("You are disgusting and beautiful")
```

```python
cla("hello nigga")
```

```python
cla1 = pipeline("zero-shot-classification")
cla1("Hi,I'm rithvik,I work with HTML",candidate_labels =["Machine learning","deep learning","Web development"],)
```

```python
generator = pipeline("text-generation")
generator("In our class,Thiruvarul is the tallest guy and")
```

```python
generator = pipeline("text-generation",model = "distilgpt2")
generator("In our class,Thiruvarul is the tallest guy and",max_length = 30,num_return_sequences =3,)
```

```python
unmasker = pipeline("fill-mask")
unmasker("<mask> game will teach you how to kick the ball and save the goal ",top_k =2)
```

```python
Ner = pipeline("ner",grouped_entities =True)
Ner("Hello,I'm Ishfaq and I work for Talibans, and I made the 911 happened")
```

```python
qaa = pipeline("question-answering")
qaa (
    question = "what is my name?",
    context = "my name is shashtri and I work in Sathyamangalam temple in sathy",
)
```
